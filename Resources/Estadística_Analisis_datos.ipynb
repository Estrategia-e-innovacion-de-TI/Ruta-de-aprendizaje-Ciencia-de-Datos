## Probabilidad y variables aleatorias

La probabilidad es un número entre 0 y 1 que expresa que tan probabile es que ocurra un evento, siendo 0 lo menos probable y 1 una certeza. Por definición es el número de eventos positivos dividido el número total de eventos, teniendo en cuenta que todos tinen la misma probabilidad de suceder. Por ejemplo, al tirar un dado la probabilidad de sacar un numero par es 3/6 = 0.5.

Ahora, cuando hablamos de eventos, nos estamos refiriendo a **variables aleatorias**. Por ejemplo, la variable aleatoria que representa al número obtenido cuándo se lanza un dado puede tomar valores entre 1 y 6.
Los valores en este rango (1-6) son llamados **espacio muestral**. 


Nosotros podemos expresar la probabilidad de que una variable aleatoria tenga un valor como *P(x=3) = 1/6* (Probabilidad de que el valor que caiga al lanzar un dado el valor sea 3)

La variable aleatoria en el ejemplo anterior es llamada **discreta**, porque tiene un **espacio muestral** que puede ser contado, es decir, son valores separados que pueden ser enumerados. 

Hay algunos casos dónde no es posible contar los valores en el **espacio muestral** porque es un rango de valores reales, estas variables son conocidas como continuas, un ejemplo de esto es el tiempo de espera mientras llega un bus al paradero.

## Distribuciones de probabilidad

En el caso de variables discretas es muy fácil expresar la probabilidad de un evento como una función P(x). Por cada valor valor *s* del espacio muestral *S* vamos a tener como probabilidad un valor entre 0 1. Es importante tener en cuenta que la suma de los valores P(x=s) para todos los valores es 1.

La distribución de probabilidad más conocida para variables discretas es la **distribución uniforme**, en la cual todos los elementos del espacio muestral *N* tienen la misma probabilidad *1/N*.

Para el caso de las variables continuas es más dificil expresar la función de probabilidad, dónde los valores pueden estar en un rango [a, b] de números reales. Esto debido a qué no existe un número finito de valores que podamos contar, por ejemplo en el caso del tiempo de espera del bus podríamos hablar de minutos, segundo, microsegundos...

Para las variables continuas, únicamente podemos hablar de la probabilidad de que el valor esté en un intervalo dentro de un rango, de esta forma P(t1≤X<t2).En este caso, la distribución de probabilidad estaría descrita por una función de densidad P(x) de la siguiente forma:

## Media, Varianza y Desviación estándar

Para comprender cada uno de los conceptos vamos a utilizar las funciones que vienen por defecto en python y las fórmulas matemáticas al momento de realizar los cálculos. Para eso utilizaremos una muestra de de 5 números

```python
# Creación de lista de números para cálculo de medidas estadísticas
numeros = [2, 4, 7, 3, 9]
```

#### Media

La media también es conocida como el promedio y es el resultado de dividir la suma de todos los valores por el total de valores (registros).

```python
media = (2 + 4 + 7 + 3 + 9)/ 5
print(media)
```

```python
# Paquete de funciones estadisticas
from statistics import mean
mean(numeros)
```

#### Varianza

La varianza es una medida de dispersión que representa la variabilidad de una serie de datos respecto a su media, es decir, que tan lejos se encuentran cada uno de los valores de la media de los mismos. Se calcula con la siguiente fórmula:

image.png

```python
varianza = (((2 - 5)**2) + ((4 - 5)**2) + ((7 - 5)**2) + ((3 - 5)**2) + ((9 - 5)**2))/(5-1)
print(varianza)
```

```python
from statistics import variance
variance(numeros)
```

#### Desviación estándar

La desviación estándar mide que tan dispersos se encuentran los datos, cuándo el valor es bajo significa que están muy concentrados y por el contrario sí el valor es alto significa que están muy dispersos. En fórmula es la raíz cuadrada de la varianza.

```python
desv_est = varianza ** (1/2)
print(desv_est)
```

```python
from statistics import stdev
stdev(numeros)
```

## Moda, Mediana y Cuartiles

```python
# Creación de lista de números para cálculo de medidas estadísticas
numeros = [2, 4, 7, 3, 9, 5, 3, 6, 4, 4]
```

#### Moda

La moda es el valor que más se repite dentro del conjunto de datos. No existe una formula para calcularlo sino que se debe revisar la frecuencia de cada uno de los valores en el total de datos.

```python
from statistics import mode
mode(numeros)
```

#### Mediana

La mediana es el valor que se encuentra en todo el centro de los datos, es el centro. A diferencia de la media, esta nos permite conocer el centro sin tener el efecto que tienen los posibles valores extremos de los datos.

```python
from statistics import median
median(numeros)
```

#### Cuartiles

Los cuartiles, son la mejor forma de entender como están distribuidos los datos que tenemos para el análisis. Por su nombre podemos entender que son 4 y consiste dividir la información en estos "espacios". Para poder hacer esta división los datos deben estar ordenados. En la siguiente figura podemos visualizar la separación:

image.png

Los cuartiles nos permiten identificar la posición de un valor respecto al conjunto total de datos. Por ejemplo, si un valor se encuentra en sobre el Q4 significa que es mayor al 75% de los datos del conjunto, por el contrario, estando en el Q1 podemos decir que es menor que el 75% de los datos del conjunto.

Existe otro concepto asociado a los cuartiles y es el **IQR (Rango intercuartil)** que es la resta entre el Q3 - Q1, que nos permite decir que los valores afuera de este rango pueden ser considerados extremos (outliers).

## Distribución Normal

Cuándo enfrentamos datos reales de cualquier tipo de medida es muy común encontrar que la distribución de los datos es muy similar, independiente de la variable que se esté mirando. Esta distribución es conocida como la **distribución normal** y tiene un rol muy importante en la estadística.

La gráfica de la distribución normal es de la siguiente forma:

image.png

En esta podemos identificar dónde se encuentran que tan concentrados están los datos y cuáles son los valores extremos mínimos y máximos, también se puede ver dónde están la media y que tan dispersos están los datos alrededor de ellos.

Utilizar una distribución normal para generar valores aleatorios, por ejemplo, para la altura de los hombres en Colombia, es la forma correcta. Conociendo la media y la desviación estándar python nos permite crear datos aleatorios para la generación de esta información, como lo mostramos a continuación:

```python
import numpy as np

media = 175
desv_est = 10
# Generando 1000 valores aleatorios en una distribución estándar con media de 175 cms
# y desviación estándar de 10 cm
muestra = np.random.normal(media,desv_est,1000)
```

```python
import matplotlib.pyplot as plt

plot = plt.hist(muestra)
```

## Covarianza y Correlación

Uno de las principales actividades que se realizan los científicos de datos es identificar relaciones entre los datos, estos son explicados mediante la correlación y la covarianza, conceptos que permiten tomar decisiones al momento de organizar la información, posiblemente uniendo o eliminando variables.

#### Correlación

La correlación es una medida estadística que explica hasta que punto unas variables están relacionadas linealmente. Sin embargo, es importante tener en cuenta que esto NO significa tengan una relación de causa y efecto.

Podemos decir que dos variables están **correlacionadas** cuándo muestran un comportamiento similar al mismo tiempo, puede ser que crezcan o disminuyan simultaneamente, en otras palabras, aparentemente existe una relación entre las variables.

La correlación es expresada en un rango entre -1 y 1, significando 0 que no tienen una correlación y mientras más cercano se encuentre el valor a 1 (ó -1) es más fuerte la correlación. 1 indicando una relación directa o positiva y -1 una relación indirecta o negativa.

# Ejercicio de conceptos aprendidos

Para poner en práctica los conceptos aprendidos, vamos a utilizar la tabla de datos de todos los jugadores que se encuentran en el videojuego FIFA 23.

```python
#Importando librerías útiles para el análisis
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leyendo el archivo con información de jugadores de FIFA23
fifa = pd.read_csv('https://raw.githubusercontent.com/JoseMVergara/Ruta_de_aprendizaje_ML/main/Fifa%2023%20Players%20Data.csv')
```

```python
#Columnas disponibles en el dataset
fifa.columns
```

```python
# Seleccionando las columnas a utilizar para el análisis
fifa = fifa[['Known As', 'Overall', 'Value(in Euro)', 'Positions Played','Nationality', 'Age', 'Height(in cm)', 'Weight(in kg)', 'Club Name', 'Preferred Foot']]
```

```python
#Muestra de las 2 primeras filas del dataset
fifa.head(2)
```

#### Medidas de tendencia central (Media, Varianza y Desviación estándar)

```python
print('La media de Overall para todos los jugadores es de:  ', statistics.mean(fifa['Overall']))
print('La varianza de Overall para todos los jugadores es de:  ', statistics.variance(fifa['Overall']))
print('La desviación estándar de Overall para todos los jugadores es de:  ', statistics.stdev(fifa['Overall']))
```

Cómo se puede observar en los resultados, la media de overall (puntaje para habilidades) para los jugadores es de 65.8. De la varianza podemos ver que los datos están muy dispersos y tienen una gran variabilidad. De la desviación estándar podemos intuir que los valores se encuentran muy concentrados cerca a la media.

```python
plot = plt.hist(fifa['Overall'])
```

Cómo se mencionaba en el análisis anterior se puede observar mediante la visualización del histograma de datos, existe un rango amplio dónde pueden estar los valores, viendo que hay valores de menos de 50 y mayores a 90, sin embargo, como mostraba la desviación estándar, en su mayoría están concentrados alrededor de un overall de 65.

```python
print('La mediana de Overall para todos los jugadores es de:  ', statistics.median(fifa['Overall']))
print('La moda de Overall para todos los jugadores es de:  ', statistics.mode(fifa['Overall']))
```

De la mediana y la moda podemos identificar que el valor más repetido es 65 pero el centro de los datos es 66, mostrando cierto efecto de los jugadores con menos overall sobre la media.

```python
print('Los cuantiles de Overall son: ', statistics.quantiles(fifa['Overall']), 'Esto significa que el 75% de los datos es mayor a', 
      statistics.quantiles(fifa['Overall'])[0], 
      ' \n y el 75% de los jugadores tienen un overall menor que', statistics.quantiles(fifa['Overall'])[2])
```

Cómo podemos ver los cuantiles los datos se encuentran concentrados en gran parte en un rango muy pequeño, entre 62 y 70. En el gráfico abajo se puede apreciar además que dentro de la distribución de los datos se pueden ver algunos valores extremos (anomalos) a la izquierda y varios más a la derecha, algo que tiene sentido porque estamos hablando de las habilidades de los jugadores del vídeo juego de FIFA y seguramente en este punto están representados los mejores jugadores del mundo.

```python
boxplot = plt.boxplot(fifa['Overall'], vert = False)
```

A continuación creamos una función que nos permite analizar rápidamente una columna dónde sacamos todas sus medidas de tendencia central y además las gráficas de su distribución y de su diagrama de caja para identificar otros aspectos relevantes de los datos que llegan en la columna.

```python
def medidas_disp(datos, columna):
  print('La media de',columna,'para todos los jugadores es de:  ', statistics.mean(datos[columna]))
  print('La varianza de',columna,'para todos los jugadores es de:  ', statistics.variance(datos[columna]))
  print('La desviación estándar de',columna,'para todos los jugadores es de:  ', statistics.stdev(datos[columna]))
  print('La mediana de',columna,'para todos los jugadores es de:  ', statistics.median(datos[columna]))
  print('La moda de',columna,'para todos los jugadores es de:  ', statistics.mode(datos[columna]))
  print('Los cuartiles de ',columna,'para los jugadores es de: ', statistics.quantiles(datos[columna]))
  plt.subplot(2, 1, 1)
  plt.hist(datos[columna])
  plt.subplot(2, 1, 2)
  plt.boxplot(datos[columna], vert = False)

  plt.show()
```

```python
medidas_disp(fifa, 'Age')
```

Usando la función para la columna de edad de FIFA podemos encontrar los siguientes aspectos relevantes:


*   La media de edad es de 25 y los datos tienen una variación alta, sin embargo estos se encuentran muy concentrados cerca de la media.
*   La edad que más se repite entre jugadores de fútbol es 22 años.
*   Los cuantiles muestran que la  gran concentración de futbolistas está en edades entre los 21 y 29 años, cifra que parece muy lógica de acuerdo con lo que se ve en un deporte de alto rendimiento.
*   La distribución deja ver que hay una cola más corta y una larga que deja intuir que hay más valores extremos hacia edades mayores. Adicionalmente la concentración está a la izquierda mostrando que hay más número de jugadores en ese rango de edades.
* El diagrama de cajas nos permite confirmar que existen valores extremos en el lado derecho, es decir, hay algunos jugadores que tienen edades muy grandes para lo normal.



##### Varianza y Covarianza

```python
fifa.corr()
```

Al determinar la matriz de correlación entre las variables que contiene el dataset podemos observar que entre todas las variables hay cierto grado de correlación. Llevando estos valores a una gráfica más visible como se muestra en la siguiente, se puede indentificar que las variables con mayor correlación son son peso y altura, sin embargo, también vemos que el overall y la edad tienen una correlación no despreciable con el valor del jugador.

```python
corr_map =sns.heatmap(fifa.corr(), annot=True, linewidth=.5, cmap="viridis")
```

```python
#pairplot es un tipo de gráfico que nos permite observar la relación de las variables y adicionalmente
#la distribución de cada una en la diagonal principal.
#También se puede agregar una variable nueva para análisis más profundos, por ejemplo usando color.
sns.pairplot(fifa, hue='Preferred Foot')
```

Aunque del gráfico no se puede ver algo que distinga los jugadores zurdos de los derechos, esta gráfica puede ser muy útil para encontrar relaciones adicionales con variables categóricas.

```python
fifa.cov()
```

La matriz de cobarianza nos muestra que hay algunas variables que sí tienen ese efecto, sin embargo, por las magnitudes que muestra la matriz es dificil identificar bien cuáles tienen ese impacto. Para solucionar esto realizamos un proceso de normalización dónde todos los valores se llevan a estar entre un rango de 1 y 0, de acuerdo con esto podemos ver más claramente el impacto.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
cov_map =sns.heatmap(np.round(scaler.fit_transform(fifa.cov())), annot=True, linewidth=.5, cmap="viridis")
cov_map.set_xticklabels(fifa.cov().columns, rotation = 90)
cov_map.set_yticklabels(fifa.cov().columns, rotation = 0)
cov_map
```

Luego de la operación de normalización se puede observar que únicamente el valor del jugador tiene covarianza significativa con todas las variables, es decir, el precio de mercado se puede mover de acuerdo a la edad, la altura, el peso y el overall de habilidades del jugador.
